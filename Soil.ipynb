##For my assignment per the "Plotting basics" module, I'm going to be organizing and visualizing data on soil moisture/temperature

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import datetime as dt
#Here, we are importing all the necessary libaries
```

```
soil = pd.read_csv("https://raw.githubusercontent.com/b-to/matplotlib_assignment/master/soilmoisture_dataset.csv")
#Here, we are importing the data from a csv file we have chosen
```

```
soil
#This is our dataframe without any modifications
#As we can see, soil values are presented for every two minute interval within each day
```

```
soil['datetime'] = pd.to_datetime(soil['datetime'])
soil['date'] = soil['datetime'].dt.date
soil['time'] = soil['datetime'].dt.time
#We're going to separate the dates and timestamps into their own separate columns to correlate each with soil moisture and soil temperate later
```

```
soil
#Now our data frame has a date column as well as a time column
```

```
soil = soil[['date', 'time', 'soil_moisture', 'soil_temperature']]
#Because we are only interested in the date, time, soil moisture, and soil temperature values within our data set, we're going to isolate those columns
```

```
soil
#Here is our data frame with the dates and timestamps separated
```

```
soil_avg = soil.groupby(['date']).mean()
#Here, we're averaging out the soil moisture and temperature values by day into a separate timeframe
```

```
soil_avg
#Here is our simplified data frame
```

```
plt.subplot(1, 2, 1)
plt.plot(soil_avg['soil_moisture'], color='red')
plt.ylabel('Moisture %')
plt.title('Soil Moisture')

plt.subplot(1, 2, 2)
plt.plot(soil_avg['soil_temperature'], color='blue')
plt.ylabel('Degrees in C')
plt.title('Soil Temperature')

#Here, we've created two line graphs showing changes in soil moisture and temperature for each day in our data fram.
#There are clearly significant changes in moisture and temperature between each day.
```

```
plt.plot(soil_avg)
plt.plot(soil_avg['soil_moisture'], color='red')
plt.plot(soil_avg['soil_temperature'], color='blue')
plt.show()
#In the line graph below, we've combined the two line graphs above to better understand the relationship between soil moisture and temperature.
#Just by looking at the graph we can reasonably assume that moisture and temperature are negatively correlated; when soil moisture is high, soil temperature is low, and vice versa.
```

```
soil.plot(kind='scatter', x='soil_moisture',y='soil_temperature', color='purple')
#To get a better understanding of the relationship between moisture and temperature, we have a scatterplot graph below.
#Looking at the plot, our previous assumption may be challenged; there are exceptions where soil temperature is low despite low soil moisture and vice versa.  
#In fact, the lowest recorded temperature value corrlates to only a moderate level of moisture.
```

```
soil.plot(kind='line', x='time',y='soil_moisture', color='red')
soil.plot(kind='line', x='time',y='soil_temperature', color='blue')
plt.show()
#Below, we have graphs that correlate the timestamps for every day in our dataframe to soil moisture and soil temperature.
#The purpose of these graphs is to get an idea of moisture and temperature consistency or lack thereof for a specific timeframe of any day.
#The soil moisture curve has a pretty consistent downwards scope as time passes in any given day.
#However, the soil temperature curve has a much less consistent upwards trend.
#This suggests that there is something else in the soil besides moisture that significantly changes temperature levels.
```
